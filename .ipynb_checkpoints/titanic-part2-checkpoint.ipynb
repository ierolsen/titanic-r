```r
# import data
train <- read.csv("data//train.csv")
test <- read.csv("data//test.csv")
```

```r
# number of gender
summary(train$Sex)
```

```r
# create a dummy survived column for the test set
test$Survived <- 0
```

```r
test$Sex == 'female'
```

```r
# our prediction is all females will survive
test$Survived[test$Sex == 'female'] <- 1
```

```r
head(test)
```

```r
summary(train$Age)
```

```r
# create a new variable 
train$Child <- 0
```

```r
train$Child[train$Age < 18] <- 1
```

```r
# number of survivors
aggregate(Survived ~ Child + Sex, data=train, FUN=sum)
```

```r
# total number of people in each subset
aggregate(Survived ~ Child + Sex, data=train, FUN=length)
```

```r
# proportion of survived or not 
aggregate(Survived ~ Child + Sex, data=train, FUN=function(x) {sum(x)/length(x)})
```

```r
# new veriable
train$Fare2 <- '30+'
```

```r
# new veriable
train$Fare2[train$Fare < 30 & train$Fare >= 20] <- '20-30'
```

```r
# new veriable
train$Fare2[train$Fare < 20 & train$Fare >= 10] <- '10-20'
```

```r
# new veriable
train$Fare2[train$Fare < 10] <- '<10'
```

```r
# aggregate new veriable
aggregate(Survived ~ Fare2 + Pclass + Sex, data=train, FUN=function(x) {sum(x)/length(x)})
```

```r
# new prediction is all people will die
test$Survived <- 0
```

```r
# all females will survive
test$Survived[test$Sex == 'female'] <- 1
```

```r
# females who have 3rd class will die
test$Survived[test$Sex == 'female' & test$Pclass == 3 & test$Fare >= 20] <- 0
```

```r
# output
submit <- data.frame(PassengerId = test$PassengerId, Survived = test$Survived)
write.csv(submit, file = "data/preds/part_2.csv", row.names = FALSE)
```
