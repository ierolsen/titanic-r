# Part 3

```r
# import data
train <- read.csv("data/train.csv")
test <- read.csv("data/test.csv")
```

```r
library(rpart)
```

```r
# making decision tree
fit <- rpart(Survived ~ Pclass + Sex + Age + SibSp + Parch + Fare + Embarked,
               data=train,
               method="class")
```

```r
# dendrogram:
plot(fit)
text(fit)
```

```r
# install packages
install.packages('rattle')
install.packages('rpart.plot')
install.packages('RColorBrewer')
```

```r
library(rattle)
library(rpart.plot)
library(RColorBrewer)
```

```r
# making prediction
Prediction <- predict(fit, test, type = "class")
```

```r
# output for decision tree
submit <- data.frame(PassengerId = test$PassengerId, Survived = Prediction)
write.csv(submit, file = "data/preds/part_3_dtree.csv", row.names = FALSE)
```

```r
# extensive decision tree
fit <- rpart(Survived ~ Pclass + Sex + Age + SibSp + Parch + Fare + Embarked,
               data=train,
               method="class", 
               control=rpart.control(minsplit=2, cp=0))
```

```r
# I will make a predict again for extensive decision tree
Prediction <- predict(fit, test, type = "class")
```

```r
# output file:
submit <- data.frame(PassengerId = test$PassengerId, Survived = Prediction)
write.csv(submit, file = "data/preds/part_3_full.csv", row.names = FALSE)
```
